```python
def invert(grid):
    """
    Take an input grid and returns a new grid of the same size and 
    shape with colors of the inverse of the input grid.
    """
    
    new_grid = grid.copy()
    # new_grid = BlockGrid(width = grid.width, height = grid.height)
    
    for block in grid:
        new_block = new_grid[block.row, block.col]
        new_block.red = 255 - block.red
        new_block.green = 255 - block.green
        new_block.blue = 255 - block.blue
     
    return new_grid
    
```
